```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("results_2.csv")
df_rbf = df[df["kernel"] == "rbf"]
df_lap = df[df["kernel"] == "laplacian"]
df_rbf.head()
```

```python
plt.plot(df_rbf["lambda"][1:], df_rbf["cce"][1:],label="CCE, RBF", linewidth=4)
plt.plot(df_lap["lambda"][1:], df_lap["cce"][1:],label="CCE, Laplacian")
plt.plot(df_rbf["lambda"][1:], df_rbf["ece"][1:],label="ECE, RBF", linewidth=4)
plt.plot(df_lap["lambda"][1:], df_lap["ece"][1:],label="ECE, Laplacian")

plt.title("ECE vs CCE for RBF and Laplacian Kernels")
plt.xlabel(r"$\lambda$")
plt.ylabel("Error")
plt.legend()
plt.show()
```

```python
plt.plot(df_rbf["lambda"][1:], df_rbf["mae"][1:],label="RBF", linewidth=4)
plt.plot(df_lap["lambda"][1:], df_lap["mae"][1:],label="Laplacian")

plt.title("MAE vs Lambda")
plt.xlabel(r"$\lambda$")
plt.ylabel("MAE")
plt.legend()
plt.show()
```

```python
plt.plot(df_rbf["lambda"][1:], df_rbf["nll"][1:],label="RBF", linewidth=4)
plt.plot(df_lap["lambda"][1:], df_lap["nll"][1:],label="Laplacian")

plt.title("NLL vs Lambda")
plt.xlabel(r"$\lambda$")
plt.ylabel("MAE")
plt.legend()
plt.show()
```
